# Distributions of vector component of Data Set 1

Here in this code combined the two files of dist1_500_1.txt and dist1_500_2.txt and generated the 10 random samples, plotted the histogram plot to those samples. Calculated the min and max for uniform distribution.


```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
import pandas as pd
from scipy import stats

data1 = pd.read_csv("dist1_500_1.txt",sep = " ",header=None).dropna() #reading the file dist1_500_1.txt
data2= pd.read_csv("dist1_500_2.txt",sep = " ",header=None).dropna()  #reading the file dist1_500_2.txt
df_list = [data1,data2] 

dataSet1= pd.concat(df_list,ignore_index=True) #combined two files 
sampleTenIndex=dataSet1.sample(10)  #generating the random ten samples

for i in range(10):
    print("index value is ",sampleTenIndex.index[i])
    data = dataSet1.iloc[sampleTenIndex.index[i]]
    print('min=%.3f max=%.3f'% (np.min(data), np.max(data)))
    plt.hist(data,bins=100)    #plotting using histogram
    plt.show()
```
