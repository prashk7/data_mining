# Distributions of vector component of Data Set 2
Here in this code combined the two files of dist2_500_1.txt and dist2_500_2.txt and generated the 10 random samples, plotted the histogram plot to those samples. Calculated the mean and standard deviation for Gaussian(Normal) distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
import pandas as pd
from scipy import stats

data1 = pd.read_csv("dist2_500_1.txt",sep = " ",header=None).dropna() #reading the file dist2_500_1.txt
data2= pd.read_csv("dist2_500_2.txt",sep = " ",header=None).dropna()  #reading the file dist1_500_2.txt
df_list = [data1,data2]

dataSet2= pd.concat(df_list)  #combined the two files 
sampleTenIndex=dataSet2.sample(10)  #generating the random ten samples

for i in range(10):
    print("index value is ",sampleTenIndex.index[i])
    data = dataSet2.iloc[sampleTenIndex.index[i]]
    print('mean=%.3f stdv=%.3f' % (np.mean(data),np.std(data)))
    plt.hist(data,bins=10)  #plotting using histogram
    plt.show()
```
